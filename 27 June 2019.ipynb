#### List Comprehensions :

```python
# N natural numbers in a list
n=10
li=[]
for i in range(1,n+1):
    li.append(i)
li=[i for i in range(1,n+1)]
li
```

```python
# Apply list comprehension to store the cubes on n natural numbers
li = [i**3 for i in range(1,11)]
li
```

```python
# Function to calculate the m factorial 
def factorail(n):
    if n == 0 or n == 1:
        return 1
    return n*factorail(n-1)
li1=[ factorail(i) for i in range(14+1)]
li1
```

```python
# Store cumulative sum of numbers till n in range
n=5
li2=[ sum(range(1,i+1)) for i in range(1,n+1)]
li2
```

```python
 #List Comprehension Function Store only leap year in a time period year%4 == 0 and year%100!=0
st = 1970
et = 2019

leapYears=[i for i in range(st,et+1)
            if i%400==0 or (i%100!=0 and i%4==0)]
leapYears
```
